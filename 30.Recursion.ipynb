```python
def calculatFactorial(num):
    if num == 0:
        return 1
    else:
        return num * calculatFactorial(num - 1)
```

```python
calculatFactorial(5)
```

```python
calculatFactorial(6)
```

```python
def calculateContigousSum(num):
    if num == 0:
        return 0
    else:
        return num + calculateContigousSum(num - 1)
```

```python
calculateContigousSum(6)
```
