```python
my_new_dictionary = {"Lewis":[40,30,np.nan],"Max":[20,np.nan,50],"Yusuf_Ali":[30,50,40], "Rob":[45,np.nan,np.nan]}
```

```python
my_new_data_frame = pd.DataFrame(my_new_dictionary)
```

```python
my_new_data_frame
```

```python
#thresh ----> verdigin degerin altinda kalan birsey varsa siler
my_new_data_frame.dropna(axis=1, thresh = 2)
```

```python
my_new_data_frame
```

```python
# fillna ----> not avaliable olanlari doldur demek
my_new_data_frame.fillna(-1)
```

```python
new_salary_frame = pd.DataFrame({"Name":["Max","Alonso","Ali"], "Salary":[10,20,30], "Age":[60,65,70]})
```

```python
new_salary_frame
```

```python
# unique ---> benzersiz
new_salary_frame["Name"].unique()
```

```python
# nuniqe ---> number of unique
new_salary_frame["Name"].nunique()
```

```python
def grossToNet(salary):
    return salary * 0.65
```

```python
# apply ----> uygula
# net maasi bulmak icin ornek
new_salary_frame["Salary"].apply(grossToNet)
```
