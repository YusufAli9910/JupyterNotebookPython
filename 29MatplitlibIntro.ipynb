```python
age_list = [20,25,30,35,40,45,50,55,60]
weight_list = [70,75,73,85,89,90,75,84,90]
```

```python
age_list_numpy = np.array(age_list)
weight_list_numpy = np.array(weight_list)
```

```python
age_list_numpy
```

```python
weight_list_numpy
```

```python
# plot --- ciz anlamina geliyor
plt.plot(age_list_numpy,weight_list_numpy, "y")
# show demem lazim gostermes icin
plt.xlabel("Age")
plt.ylabel("Wight")
plt.title("Age-Weight graph")
plt.show()
```

```python
my_numpy1 = np.linspace(0,15,20)
```

```python
my_numpy1
```

```python
my_numpy2 = my_numpy1 ** 3
```

```python
my_numpy2
```

```python
plt.plot(my_numpy1,my_numpy2, "g-*")
plt.show()
```

```python
# subplot ----> alt grafik
plt.subplot(1,2,1) # 1 row, 2 column, 1.graph
plt.plot(my_numpy1,my_numpy2,"r--")
plt.subplot(1,2,2)
plt.plot(my_numpy2,my_numpy1,"g*")
plt.show()
```

```python
my_figure = plt.figure()
# [left, bottom, width, height]
my_axes = my_figure.add_axes([1,1,0.5,0.5])
my_axes.plot(my_numpy1,my_numpy2, "g*-")
my_axes.set_xlabel("X Data")
my_axes.set_ylabel("Y Data")
my_axes.set_title("The graph")
plt.show()
```

```python
my_figure2 = plt.figure()
my_axes3 = my_figure2.add_axes([0.3,0.3,0.9,0.9])
my_axes3.plot(my_numpy1,my_numpy2, "r*-")
my_axes3.set_xlabel("X Data Large")
my_axes3.set_ylabel("Y Data Large ")
my_axes3.set_title("Large Graph")



# [left, bottom, width, height]
my_axes2 = my_figure2.add_axes([0.4,0.6,0.3,0.3])
my_axes2.plot(my_numpy1,my_numpy2, "g*-")
my_axes2.set_xlabel("X Data Small")
my_axes2.set_ylabel("Y Data Small ")
my_axes2.set_title("Small Graph")


plt.show()
```

```python
#(my_fig, my_axe --> biz figur ve axis dondutuyor bu metod)
(my_fig, my_axe) = plt.subplots()
my_numpy3 = np.linspace(0,15,20)
my_numpy4 = my_numpy3 ** 2
my_axe.plot(my_numpy3,my_numpy4, color="#060270", alpha=1)
my_axe.plot(my_numpy4,my_numpy3, color = "#CC6CE7", alpha =0.9)
# alpha ile gorunus ve saydamlik tonunu ayarliyoruz.
plt.show()
```

```python
(new_fig, new_axe) = plt.subplots()
new_axe.plot(my_numpy3, my_numpy3 + 2, color="#751A8F", linewidth = 6)
new_axe.plot(my_numpy3, my_numpy3 + 3, color="#14DCD9", linewidth = 3)
new_axe.plot(my_numpy3, my_numpy3 + 4, color="red", linewidth = 0.8, linestyle="-.")
new_axe.plot(my_numpy3, my_numpy3 + 5, color="g", linewidth = 2, linestyle = ":")
new_axe.plot(my_numpy3, my_numpy3 + 6, color="pink", linewidth = 2, linestyle = "--", marker = "o")
new_axe.plot(my_numpy3, my_numpy3 + 7, color="black", linewidth = 2, linestyle = "--", marker = "+", markersize = 9)

# marker ----> pointleri gosterir cizgide.
# linestyle ---> cizgi tasarimi vs.
# linewidth ise ---> cizgi kalinligi
plt.show()
```

```python
# scatter ---> dagitmak
plt.scatter(my_numpy3,my_numpy4)
plt.show()
```
